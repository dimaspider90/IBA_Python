```python
# 1. Написать алгоритмы сортировок insert и shaker по псевдокоду
# 2.Вставить их в код выше, вывести время выполнения сортировки(как для пузырька и быстрой сортирововок) и добавить результат в график
# 3. Добавьте заголовок графика и подписи для осей
# 4. Создайте точечный график для сортировок 
# 5. Создайте BarPlot для сортировок 
# 6. Разделите сортировки на отдельные графики




```

```python
import random
import datetime
import matplotlib.pyplot as plt

# сортировка пузырьком
def BubbleSort(A):
    for i in range(len(A)):
        for j in range (len(A)-1-i):
            if A[j] > A[j+1]:
                temp = A[j]
                A[j] = A[j+1] 
                A[j+1] = temp
                
def insertionSort(B):
    for i in range(1, len(B)):
        t = B[i]
        j = i - 1
        while j >= 0 and t < A[j]:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = t

def shakerSort(C):
    for i in range (0, N // 2):
        for j in range(0, N-1-i):
            if C[j]>C[j+1]:
                temp = C[j]
                C[j] = C[j+1]
                C[j+1] = temp
             
def quickSort(D):
   if len(D) <= 1:
       return D
   else:
       q = random.choice(D)
       s_nums = []
       m_nums = []
       e_nums = []
       for n in D:
           if n < q:
               s_nums.append(n)
           elif n > q:
               m_nums.append(n)
           else:
               e_nums.append(n)
       return quickSort(s_nums) + e_nums + quickSort(m_nums)
        
x = []
y1_bubble = []
y2_insert = []
y3_shaker = []
y4_quick = []

for N in range(1000,5001,1000):
    x.append(N)
    min = 1
    max = N
    A = []
    for i in range (N):
        A.append(int(round(random.randint(0, 1000))))
    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1_bubble.append((t2 - t1).total_seconds())
    print("Пузырьковая сортировка   " +str(N)+"   заняла   "+str((t2-t1).total_seconds()) + "c")
    
# Делаем копию списка B
    B = A.copy()
    
    t3 = datetime.datetime.now()
    insertionSort(B)
    t4 = datetime.datetime.now()
    y2_insert.append((t4 - t3).total_seconds())
    print("Cортировка вставки  " +str(N)+"   заняла   "+str((t4-t3).total_seconds()) + "c")

# Делаем копию списка C
    C = A.copy()
    
#    print (C)
    t5 = datetime.datetime.now()
    shakerSort(C)
    t6 = datetime.datetime.now()
    y3_shaker.append((t6 - t5).total_seconds())
    print("Cортировка шейкером  " +str(N)+"   заняла   "+str((t6-t5).total_seconds()) + "c")
    
    # Делаем копию списка D
    D = A.copy()
    
#    print (D)
    t7 = datetime.datetime.now()
    quickSort(D)
    t8 = datetime.datetime.now()
    y4_quick.append((t8 - t7).total_seconds())
    print("Cортировка быстрая " +str(N)+"   заняла   "+str((t8-t7).total_seconds()) + "c")
    print("---")
    
# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x, y1_bubble, "C0", label="Bubble Sort")
plt.plot(x, y2_insert, "C1", label="Insert Sort")
plt.plot(x, y3_shaker, "C2", label="Shaker Sort")
plt.plot(x, y4_quick, "C3", label="Quick Sort")
plt.title("Сравнение времени выполнения сортировок")
plt.xlabel("Размер списка")
plt.ylabel("Время (секунды)")
plt.legend()
plt.grid(True)
plt.show()

# BarPlot для сортировок
plt.figure(figsize=(10, 6))
width = 200
plt.bar([p - width for p in x], y1_bubble, width=width, label="Bubble Sort")
plt.bar(x, y2_insert, width=width, label="Insert Sort")
plt.bar([p + width for p in x], y3_shaker, width=width, label="Shaker Sort")
plt.bar([p + 2 * width for p in x], y4_quick, width=width, label="Quick Sort")
plt.title("Сравнение времени выполнения сортировок (BarPlot)")
plt.xlabel("Размер списка")
plt.ylabel("Время (секунды)")
plt.legend()
plt.grid(True)
plt.show()

# Разделение сортировок на отдельные графики
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y1_bubble, "C0")
axs[0, 0].set_title("Bubble Sort")
axs[0, 0].set_xlabel("Размер списка")
axs[0, 0].set_ylabel("Время (секунды)")

axs[0, 1].plot(x, y2_insert, "C1")
axs[0, 1].set_title("Insert Sort")
axs[0, 1].set_xlabel("Размер списка")
axs[0, 1].set_ylabel("Время (секунды)")

axs[1, 0].plot(x, y3_shaker, "C2")
axs[1, 0].set_title("Shaker Sort")
axs[1, 0].set_xlabel("Размер списка")
axs[1, 0].set_ylabel("Время (секунды)")

axs[1, 1].plot(x, y4_quick, "C3")
axs[1, 1].set_title("Quick Sort")
axs[1, 1].set_xlabel("Размер списка")
axs[1, 1].set_ylabel("Время (секунды)")

plt.tight_layout()
plt.show()
```
